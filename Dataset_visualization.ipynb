**Our Propsed Urdu Dataset(UD23)**

```python
import pickle
Preprocessed_Urdu_Data_for_training= pd.read_pickle("/content/drive/MyDrive/Colab Notebooks/Dataset/Processed_Dataset/Preprocessed_Urdu_Data_for_training.pkl")
```

```python
Preprocessed_Urdu_Data_for_training.info()
```

```python
Preprocessed_Urdu_Data_for_training.describe()
```

```python
# Drop any rows with missing values
Preprocessed_Urdu_Data_for_training.dropna(subset=['Original_Review'], inplace=True)

# Split list into new series
lyrics = Preprocessed_Urdu_Data_for_training['Original_Review'].str.lower().str.split()

# Get amount of unique words
Preprocessed_Urdu_Data_for_training['Original_Review_Unique_words_Counter'] = lyrics.apply(set).apply(len)

# Get amount of words
Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'] = lyrics.apply(len)
```

```python
Preprocessed_Urdu_Data_for_training
```

```python
ax=sns.countplot(x='Sentiment', data=Preprocessed_Urdu_Data_for_training)
# add count labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
```

```python
# find the maximum value in column 'A'
max_Lengyth_Original_review = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].max()

# print the maximum value
print("Max length of original review words are")
print(max_Lengyth_Original_review)
```

```python
# find the maximum value in column 'A'
min_Lengyth_Original_review = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].min()

# print the maximum value
print("Min length of original review words are")
print(min_Lengyth_Original_review)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create a distribution plot of the 'Original_Review_Words_Counter' column
sns.histplot(Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'])

# get the maximum and minimum values of the column
max_value = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].max()
min_value = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].min()

# add vertical lines to the plot to show the maximum and minimum values
plt.axvline(max_value, color='red', linestyle='--', linewidth=2.5)
plt.axvline(min_value, color='green', linestyle='--', linewidth=2.5)

# add labels to the vertical lines
plt.text(max_value, 0.05, f'Maximum: {max_value:.0f}', color='red', fontsize=12, va='baseline', ha='right')
plt.text(min_value, 0.05, f'Minimum: {min_value:.0f}', color='green', fontsize=12, va='bottom', ha='left')


# show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# get the minimum and maximum values
min_val = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].min()
max_val = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].max()

# create a bar plot
plt.bar(['Minimum', 'Maximum'], [min_val, max_val])

# annotate the values on the bars
plt.text(0, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')
plt.text(1, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')

# set the title and axis labels
plt.title('Minimum and Maximum length of Original Review')
plt.ylabel('Value')

# show the plot
plt.show()


```

**Maximum and Minimum Length of Urdu reviews**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create the violin plot
sns.violinplot(y=Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'])

# get the minimum and maximum values
min_val = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].min()
max_val = Preprocessed_Urdu_Data_for_training['Original_Review_Words_Counter'].max()


# add annotations for the minimum and maximum values
plt.annotate(f'Minimum: {min_val:.0f}', xy=(0, min_val), xytext=(80, 10),
             textcoords='offset pixels', va='bottom', ha='left', fontsize=12, color='red')
plt.annotate(f'Maximum: {max_val:.0f}', xy=(0, max_val), xytext=(10, 0),
             textcoords='offset points', va='center', ha='left', fontsize=12, color='green')

# set the title and axis labels
plt.title('Minimum and Maximum Length of Reviews')
plt.ylabel('Length of Reviews')

# show the plot
plt.show()
```
