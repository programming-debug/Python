```python
import pandas as pd
import seaborn as sns
```

```python
import pickle
underSample_UDSA_23 = pd.read_pickle("/content/drive/MyDrive/Colab Notebooks/Dataset/Processed_Dataset/balanced_dataset.pkl")
```

```python
import matplotlib.pyplot as plt
ax=sns.countplot(x='Sentiment', data=underSample_UDSA_23)
# add count labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 6), 
                textcoords = 'offset points')
    
ax.set(title="Distribution of Data")
# Set the x-axis and y-axis labels
ax.set(xlabel="Type of Sentiment Label", ylabel="Number of Reviews")
plt.savefig('/content/drive/MyDrive/Extracted/underSample_UDSA_23.pdf', dpi=300, bbox_inches='tight')
plt.show()
# Save the plot as a PNG file
```
