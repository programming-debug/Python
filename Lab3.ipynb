```python
class Node:
  def __init__(self,state,parent,actions,totalCost):
    self.state=state
    self.parent=parent
    self.actions=actions
    self.totalCost=totalCost
    
```

This code defines a Python class called "Node". Each instance of this class represents a node in a graph or tree data structure. The node has four attributes:

**"state":** This is the state or value associated with the node. The meaning of this attribute depends on the context of the graph or tree. For example, in a graph representing a maze, the state might be the coordinates of a cell in the maze.

**"parent":** This is a reference to the parent node of the current node. This attribute allows us to traverse the graph or tree in reverse order from a leaf node to the root.

**"actions":** This is a list of actions or edges that lead from the parent node to the current node. In other words, if we apply each action in this list starting from the parent node, we will arrive at the current node.

**"totalCost":** This is the total cost or distance associated with the path from the root node to the current node. The meaning of this attribute also depends on the context of the graph or tree. For example, in a graph representing a transportation network, the total cost might be the sum of the travel times along each edge in the path.

```python
graph={
    'A':Node('A',None,['B','C','E'],None),
    'B':Node('B',None,['A','D','E'],None),
    'C':Node('C',None,['A','F','G'],None),
    'D':Node('D',None,['B','E'],None),
    'E':Node('E',None,['A','B','D'],None),
    'F':Node('F',None,['C'],None),
    'G':Node('G',None,['C'],None)
}
```

This is a Python code defining a dictionary called "graph", which represents a simple undirected graph using an adjacency list representation.

Adjacency list representation is a way to represent a graph using a list of lists or a dictionary of lists.


Each key in the dictionary represents a vertex in the graph, and its corresponding value is a Node object, which represents the state of the vertex in a search problem.


For example, the vertex 'A' is represented by a Node object with state 'A', a parent of None, a list of actions ['B', 'C', 'E'], and a total cost of None. This means that vertex 'A' has edges to vertices 'B', 'C', and 'E' in the graph. Similarly, the other vertices are represented by their corresponding Node objects.

Note that the parent and total cost fields in the Node objects are not initialized in this code. These fields would typically be set by a search algorithm during the search process.

Overall, this code provides a simple way to define a graph and its vertices, which can be used as a starting point for many graph-based algorithms.

**Activity 2:**

For the graph in previous activity, imagine **node A** as **starting node** and your **goal** is to reach **F**. Keeping breadth first search in mind, describe a sequence of actions that you must take to reach that goal state

```python
def actionSequence(graph,initialState,goalState):
  solution=[goalState]
  currentParent=graph[goalState].parent
  while currentParent!=None:
    solution.append(currentParent)
    currentParent=graph[currentParent].parent
  solution.reverse()
  return solution
```

```python
def BFS():
  initialState='A'
  goalState='F'
  graph={
    'A':Node('A',None,['B','C','E'],None),
    'B':Node('B',None,['A','D','E'],None),
    'C':Node('C',None,['A','F','G'],None),
    'D':Node('D',None,['B','E'],None),
    'E':Node('E',None,['A','B','D'],None),
    'F':Node('F',None,['C'],None),
    'G':Node('G',None,['C'],None)}
  frontier=[initialState]
  explored=[]
  while len(frontier)!=0:
    currentNode= frontier.pop(0)
    explored.append(currentNode)
    for child in graph[currentNode].actions:
      if child not in frontier and child not in explored:
        graph[child].parent=currentNode
        if graph[child].state==goalState:
          return actionSequence(graph, initialState,goalState)
        frontier.append(child)

```

```python
solution=BFS()
print(solution)
```
